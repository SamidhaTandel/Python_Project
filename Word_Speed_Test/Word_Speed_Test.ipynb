```python
from time import *
import random as r


#Mistake function will return no of errors if any index word is missed 
def mistake(par,typetest):
    error=0
    for i in range(len(par)):
        try:    
            '''
        it checks wether string index which is given for the text is
        equal to the string index which is written by user and error
        varinable will increase the value by 1 
        
            '''
            if par[i]!=typetest[i]: 
                error=error+1   
        except:
            error=error+1
    return error

# speed_time fucntion will return the speed of the word/sec 
def speed_time(time_s,time_e,userinput):
    time_delay=time_e-time_s #diffrence between starttime and endtime
    time_R=round(time_delay,2)
    speed=len(userinput)/time_R #length of userinput i.e upon difference of time
    return round(speed)

#Create variable to get name in your word_speed_test
name=input("Enter your name : ")
name_str='My name is : '

#Creating list of paragraphs for typing test
word_lst=["0 (zero) is a number representing an empty quantity.",
          "Adding 0 to any number leaves that number unchanged.",
         "Aryabhatta, the famous Indian mathematician discovered 'zero'.",
          name_str+name]

#Creating test1 which will give random string from the list for typing test
test1= r.choice(word_lst)

print()
print()
print("---------- TYPING TEST ---------")
print()
print()
#Printing the string for the test which is stored in variable test1
print(test1)
print()
#time_start will store initial time of the user
time_start=time()
usertest=input(" Type the sentence printed to get the speed :  ")
#time_end will store the end time of the user
time_end=time()


print()
print()

print("----------RESULT---------")
print()
print()

print("Out of ",len(test1))
print("Error : ",mistake(test1,usertest))
print("Speed Test Result : ",speed_time(time_start,time_end,usertest),"words/sec")
```
